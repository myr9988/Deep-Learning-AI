```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_data_gen = ImageDataGenerator(rescale=1/255)
training_data = train_data_gen.flow_from_directory('../input/cat-and-dog/training_set/training_set',
                                                   target_size=(64,64),batch_size=32,class_mode='binary')
```

```python
test_data_gen = ImageDataGenerator(rescale=1/255)
test_data = test_data_gen.flow_from_directory('../input/cat-and-dog/test_set/test_set',
                                                   target_size=(64,64),batch_size=32,class_mode='binary')
```

# Convolutional Neural Network

```python
cnn = tf.keras.models.Sequential()
```

```python
#Convolution
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
```

```python
#Pooling
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
#Adding second layer
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
#Flattening
cnn.add(tf.keras.layers.Flatten())
```

```python
#Full connection
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
#Output layer
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
#Compile the CNN
cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
```

```python
#Running the CNN
cnn.fit(x=training_data,validation_data=test_data,epochs=5)
```

```python
import numpy as np
from keras.preprocessing import image
val_image = image.load_img('../input/cat-and-dog/test_set/test_set/dogs/dog.4018.jpg',target_size=(64,64))
val_image = image.img_to_array(val_image)
val_image = np.expand_dims(val_image,axis=0)
```

```python
result = cnn.predict(val_image)
result
```
